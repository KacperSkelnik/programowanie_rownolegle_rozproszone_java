```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
y = [round(float(line.strip()),3) for line in open("laboratory/src/main/java/lab_1/dane_do_wykresu.txt", 'r')]
x = [32, 64, 128, 256, 512, 1024, 2048, 4096, 8192]
```

```python
plt.figure(figsize=(9, 9))
plt.plot(x,y)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("pixels", size=20)
plt.ylabel("t [ms]", size=20)

plt.savefig("laboratory/src/main/java/lab_1/lab1.png")
plt.show()
```
